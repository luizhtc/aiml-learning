```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from utils import plot_cluster
```

```python
iris = datasets.load_iris()
iris.keys()
```

```python
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(iris.data)
print(agglo_labels)
results = confusion_matrix(iris.target, agglo_labels)
print(results)
```

```python
plot_cluster(iris.data, agglo_labels, 'Cluster Analysis with Agglomerative Clustering')
```

```python
plt.figure(figsize=(12, 6))
plt.title('Hierarchy Clustering: Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
linkage_matrix = linkage(iris.data, method='ward')
dendrogram(linkage_matrix, truncate_mode='lastp', p=15)
plt.axhline(y=7, c='grey', lw=1, linestyle='dashed')
plt.show()
```
